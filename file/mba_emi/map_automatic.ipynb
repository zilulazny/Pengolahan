```python
#for question : emiulfiana226@gmail.com
#BMKG Denpasar

import pandas as pd
from datetime import datetime
import pytz
import subprocess

file=pd.read_excel('nov2.xlsx',skiprows=1)
eq_info=file['Info Gempa']
eq_file1=open('eq_shallow.dat','w')
eq_file2=open('eq_med.dat','w')
eq_file3=open('eq_deep.dat','w')
eq_file_all=open('eq_all.dat','w')


for info in eq_info:
    magn=float(info.split("Mag:")[1].split(" ,")[0])
    time_str=info.split(",")[1].strip()
    time_wib=datetime.strptime(time_str, "%d-%b-%y %H:%M:%S WIB")
    time_wib = pytz.timezone('Asia/Jakarta').localize(time_wib)
    time_utc = time_wib.astimezone(pytz.UTC)
#    print(time_utc)
    loc=info.split("Lok:")[1]
    if "LS" in loc:
        longt=float(info.split("Lok:")[1].split("LS,")[1].split(" BT")[0])
        latt=-float(info.split("Lok:")[1].split("LS,")[0])
    elif "LU" in loc:
        longt=float(info.split("Lok:")[1].split("LU,")[1].split(" BT")[0])
        latt=float(info.split("Lok:")[1].split("LU,")[0])
    else:
        longt=latt="Please check the data"
        print("Please check the epicenter")
    dep=int(info.split("Kedlmn:")[1].split('Km')[0])
    
    #filter the source and time information
    #give the filter information
    magn_1 = 0
    magn_2 = 10
    time_1 = datetime(2024,11,1,0,0,0,tzinfo=pytz.UTC)
    time_2 = datetime(2024,12,1,0,0,0,tzinfo=pytz.UTC)
    dep_1 = 60
    dep_2 = 300
    lat_1 = -12
    lat_2 = -6.75
    long_1 = 113.21
    long_2 = 117.31
    
    if time_1 <= time_utc < time_2:
#        print(time_utc)
        if magn_1 <= magn <= magn_2:
            if long_1 <=longt <=long_2 and lat_1 <= latt <=lat_2:
                if 0 <= dep <= dep_1 :
                    eq_file1.write(f"{longt}\t{latt}\t{dep}\t{magn}\t{time_utc}\n")
                elif dep_1 < dep <= dep_2:
                    eq_file2.write(f"{longt}\t{latt}\t{dep}\t{magn}\t{time_utc}\n")
                elif dep > dep_2:
                    eq_file3.write(f"{longt}\t{latt}\t{dep}\t{magn}\t{time_utc}\n")
                else:
                    print("please check the hypocenter")
                eq_file_all.write(f"{longt}\t{latt}\t{dep}\t{magn}\t{time_utc}\n")

```

```python
#gmt map
gmt_script='map_bali.csh'
try:
    gmt_run=subprocess.run(['csh',gmt_script], check=True, capture_output=True, text=True)
    print("gmt_script:", gmt_run.stdout)
except subprocess.CalledProcessError as e:
    print("An error occurred while running the script:", e.stderr)
```
