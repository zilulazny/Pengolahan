```python
import numpy as np
import pygmt
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('datalengkap.csv')

# Memilih kolom yang diinginkan
selected_columns = df[['Bujur','Lintang','Kedalaman', 'Magnitude']]

# Menulis data ke file DAT
with open('inputpeta.dat', 'w') as file:
    for index, row in selected_columns.iterrows():
        line = '  '.join(row.astype(str))  # Menggabungkan semua elemen baris menjadi satu string dengan spasi sebagai pemisah
         file.write(line + '\n')
            
 # Membaca data dari hasil2.csv
data = pd.read_csv('datalengkap.csv')

# Buat figure
fig = pygmt.Figure()

# Buat color palette untuk kedalaman gempa
pygmt.makecpt(cmap="geo", series=[-7000, 4000])
pygmt.makecpt(cmap="red,yellow,green", series="0,60,300,1000", output="quakes.cpt")
    
#title_text = f"Peta Seismisitas Wilayah Bali dan Sekitarnya \nPeriode {start_date_formatted} hingga {end_date_formatted}"


# Plot peta dasar dengan topografi
fig.grdimage(
    grid="@earth_relief_15s",
    region=[113.21, 117.31, -12, -7],
    projection="M17c",
    shading=True,
    frame=["xa2g2", "ya2g2"])#'+t"Peta Seismisitas"'])

# Plot gempa yang tidak dirasakan (kolom 'Dirasakan' kosong)
gempa_tidak_dirasakan = data[data['Dirasakan'].isna()]
if not gempa_tidak_dirasakan.empty:
    fig.plot(
        x=gempa_tidak_dirasakan['Bujur'],
        y=gempa_tidak_dirasakan['Lintang'],
        size=0.08*gempa_tidak_dirasakan['Magnitude'],
        style="cc",  # Lingkaran
        fill=gempa_tidak_dirasakan['Kedalaman'].astype(float),  # Warna berdasarkan kedalaman
        cmap="quakes.cpt",  # Gunakan colormap yang telah ditentukan
        pen="black"
    )
else:
    print("Tidak ada gempa bumi yang tidak dirasakan untuk diplot.")

# Plot gempa yang dirasakan (kolom 'Dirasakan' terisi)
gempa_dirasakan = data[~data['Dirasakan'].isna()]
if not gempa_dirasakan.empty:
    fig.plot(
        x=gempa_dirasakan['Bujur'],
        y=gempa_dirasakan['Lintang'],
        size=0.08*gempa_dirasakan['Magnitude'],
        style="a0.5c",  # Simbol bintang
        fill="red",  # Warna bintang merah
        pen="1p,black"  # Garis tepi hitam dengan ketebalan 1 point
    )
else:
    print("Tidak ada gempa bumi yang dirasakan untuk diplot.")

# Plot fault lines
fig.plot(data="D:/latsar/aktualisasi/fault.gmt", pen="0.6,black")
fig.plot(data="D:/latsar/aktualisasi/trench.gmt", pen="4,black")

    
# Menambahkan colorbar
fig.image('D:/latsar/aktualisasi/coba/LogoBMKG.png', position='n1.1/0.85+w1.1i')
fig.image('D:/latsar/aktualisasi/peta/mtangn.png', position='n1.04/0.7+w1.1i')
fig.basemap(map_scale="n1.12/0.65+w80k+f0.1p+lKm")
fig.colorbar(
    frame=["x+lKetinggian", "y"],  # Label untuk sumbu x dan y
    position="n1.24/0.63+w4c/0.5c")  # Posisi colorbar (JMR = tengah-kanan), lebar 0.5 cm, panjang 10 cm
fig.legend(spec="D:/latsar/aktualisasi/coba/legend.gmt", position="n1.03/0.04+w5c",box="+gwhite+p1p,black")
with fig.inset(position="n1.03/0.28+w5c/2.5c", box="+pblack"):
    fig.coast(
        region=[97, 140, -15, 7],
        shorelines='0.5p,black',
        projection="M5c",
        land="green",
        water="lightblue"
    )
    fig.plot(x=[117.31, 113.21, 113.21, 117.31, 117.31], y=[-7, -7, -12, -12, -7], pen="1p,red")

# Membuat cross section
fig.plot(x=[114.8, 115.4], y=[-11.5, -7.5], projection="M", pen=2)
fig.text(x=114.9, y=-11.6, text="A", font="18,Helvetica")
fig.text(x=115.55, y=-7.5, text="A1", font="18,Helvetica")
pygmt.project(
    data="inputpeta.dat",
    unit=True,
    center=[114.8, -11.5],
    endpoint=[115.4, -7.5],
    width=[-200, 200],
    convention="pz",
    outfile="crsx8.dat"
)

fig.shift_origin(yshift=9, xshift=17.5)
fig.plot(
    x=[113, 113, 114.18, 114.18],  # Koordinat X dari sudut-sudut persegi
    y=[-11.4, -12, -12, -11.4],  # Koordinat Y dari sudut-sudut persegi
    fill="white",  # Warna latar belakang
    close=True  # Menutup jalur untuk membuat persegi berisi warna 
)

# Buat peta cross section dengan basemap
fig.basemap(
    projection="X4c/-2.5c",
    region=[0, 450, 0, 300],
    frame=['xafg+lDistance (km)','yafg+lDepth (km)', "wsEN"]
)

# Membuat garis cross section    
fig.plot(data="crsx.dat", projection="X", style="c0.15", pen=0.3, fill='red')
fig.text(x=20, y=25, text="A", font="11,Helvetica")
fig.text(x=370, y=25, text="A1", font="11,Helvetica")

# Menyimpan gambar
fig.savefig('ini ris.png')
#fig.show

```

```python
import pygmt
import pandas as pd

# 1. Membaca data dari CSV (sesuaikan dengan file path Anda)
df = pd.read_csv('datalengkap.csv')

# 2. Pastikan kolom-kolom yang digunakan memiliki tipe data numerik (convert jika perlu)
df['Bujur'] = pd.to_numeric(df['Bujur'], errors='coerce')  # Mengubah Bujur menjadi numerik
df['Lintang'] = pd.to_numeric(df['Lintang'], errors='coerce')  # Mengubah Lintang menjadi numerik
df['Kedalaman'] = pd.to_numeric(df['Kedalaman'], errors='coerce')  # Mengubah Kedalaman menjadi numerik

# 3. Hapus baris yang memiliki nilai NaN (jika ada)
df = df.dropna(subset=['Bujur', 'Lintang', 'Kedalaman'])

# 4. Pilih data yang diperlukan: Bujur, Lintang, dan Kedalaman
data = df[['Bujur', 'Lintang', 'Kedalaman']]

# 5. Tentukan lintasan (misalnya dari [Bujur1, Lintang1] ke [Bujur2, Lintang2])
x_start, y_start = 113.5, -11  # Titik awal lintasan
x_end, y_end = 115.5, -8  # Titik akhir lintasan

# 6. Pilih data yang berada pada lintasan ini (dengan kedalaman yang relevan)
# Tentukan rentang X dan Y untuk lintasan
cross_section_data = data[
    (data['Bujur'] >= x_start) & (data['Bujur'] <= x_end) &
    (data['Lintang'] >= y_end) & (data['Lintang'] <= y_start)
]

# 7. Buat color palette untuk kedalaman dengan 3 kategori warna
# Warna merah untuk kedalaman < 60m, kuning untuk kedalaman 60-300m, hijau untuk kedalaman > 300m
pygmt.makecpt(cmap="red,yellow,green", series="0,60,300,7000", output="custom.cpt")

# 8. Membuat figure
fig = pygmt.Figure()

# 9. Plot data pada lintasan berdasarkan kedalaman
fig.plot(
    x=cross_section_data['Bujur'], 
    y=cross_section_data['Kedalaman'], 
    style="c0.5c", 
    pen="black", 
    fill=cross_section_data['Kedalaman'], 
    cmap="custom"  # Menggunakan colormap khusus
)

# 10. Menambahkan label kedalaman untuk setiap titik
for i in range(len(cross_section_data)):
    fig.text(
        cross_section_data['Bujur'].iloc[i], 
        cross_section_data['Kedalaman'].iloc[i], 
        text=f"{cross_section_data['Kedalaman'].iloc[i]:.1f}", 
        font="8p,Helvetica-Bold,black", 
        offset=(0.2, 0.2)
    )

# 11. Menambahkan basemap untuk cross-section
fig.basemap(
    projection="X6c/-4c",  # Proyeksi cross-section
    region=[x_start, x_end, 0, 7000],  # Rentang X dan Y (0 hingga 7000 meter)
    frame=["xafg+l'Jarak (km)'", "yafg+l'Kedalaman (m)'"]
)

# 12. Menyimpan gambar
fig.savefig("cross_section_with_custom_depth_colors.png")
fig.show()
```

```python
import numpy as np
import pygmt
import pandas as pd

# Membaca data dari 'datalengkap.csv'
df = pd.read_csv('datalengkap.csv')

# Menyaring kolom yang diperlukan untuk peta seismisitas
selected_columns = df[['Bujur', 'Lintang', 'Kedalaman', 'Magnitude']]

# Menulis data ke file DAT untuk peta seismisitas
with open('inputpeta.dat', 'w') as file:
    for index, row in selected_columns.iterrows():
        line = '  '.join(row.astype(str))  # Menggabungkan semua elemen baris menjadi satu string
        file.write(line + '\n')

# Membaca kembali data untuk peta seismisitas
data = pd.read_csv('datalengkap.csv')

# Membuat figure peta dasar
fig = pygmt.Figure()

# Membuat color palette untuk kedalaman gempa
pygmt.makecpt(cmap="red,yellow,green", series="0/60/300/7000", output="custom.cpt")

# Plot peta dasar dengan topografi menggunakan grid dari @earth_relief_15s
fig.grdimage(
    grid="@earth_relief_15s", 
    region=[113, 118, -15, -6],  # Sesuaikan dengan wilayah data
    projection="M6c",  # Proyeksi Mercator
    shading=True,  # Shading untuk topografi
    frame=True
)

# Plot titik-titik gempa dengan warna berdasarkan kedalaman
fig.plot(
    x=data['Bujur'], 
    y=data['Lintang'], 
    size=0.08 * data['Magnitude'],  # Ukuran berdasarkan Magnitudo
    style="c0.5c",  # Simbol lingkaran
    pen="1p,black",  # Garis tepi hitam
    fill=data['Kedalaman'],  # Warna berdasarkan kedalaman
    cmap="custom"  # Gunakan colormap yang telah dibuat
)

# Menambahkan colorbar untuk kedalaman
fig.colorbar(frame=["x+lKedalaman", "y"])

# Menyimpan peta dasar
fig.savefig("output_peta.png")

# 2. Membuat Data Cross Section (crsx8.dat)
# Tentukan koordinat awal dan akhir untuk garis cross-section
start_lon, start_lat = 114.8, -11.5  # Titik awal garis cross-section
end_lon, end_lat = 115.4, -7.5  # Titik akhir garis cross-section

# Menyaring data berdasarkan garis cross-section
data_cross_section = data[
    (data['Bujur'] >= start_lon) & (data['Bujur'] <= end_lon) &
    (data['Lintang'] >= min(start_lat, end_lat)) & (data['Lintang'] <= max(start_lat, end_lat))
]

# Menyimpan data cross-section dalam format .dat
data_cross_section = data_cross_section[['Bujur', 'Kedalaman', 'Magnitude']]
data_cross_section.to_csv('crsx8.dat', sep=' ', index=False, header=False)

# 3. Plot Cross Section
# Membuat figure untuk cross-section
fig = pygmt.Figure()

# Tentukan proyeksi dan region untuk cross section
fig.basemap(
    projection="X4c/-2.5c", 
    region=[0, 450, 0, 7000],  # Sesuaikan range kedalaman dengan data yang ada
    frame=['xafg+l"Distance (km)"', 'yafg+l"Depth (km)"', "wsEN"]
)

# Plot data cross-section dengan warna berdasarkan kedalaman
fig.plot(
    data="crsx8.dat",  # Data file cross-section
    style="c0.15",  # Ukuran titik kecil
    pen=0.3,  # Garis tepi
    fill="2",  # Mengisi warna berdasarkan kolom Kedalaman (kolom kedua)
    cmap="custom"  # Menggunakan colormap kustom yang sudah dibuat
)

# Menambahkan colorbar untuk kedalaman
fig.colorbar(cmap="custom", frame=["x+lDistance", "y+lDepth"])

# Menyimpan gambar cross-section
fig.savefig('cross_section.png')

# Menampilkan gambar
fig.show()
```
